```python
# Import Packages

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files 
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#read train data
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
#print first 5 rows of train data 
train_data.head()
```

```python
#read train data
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
#print first 5 rows of test data 
test_data.head()
```

```python
#see percentages of survival in some fields in training data
pd.pivot_table(train_data, index = 'Survived', values = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare'])
```

Fast conclusions: People with higher Fare have higher surviving percentages......
People with lower Pclass have higher surviving percentages

```python
#compare the survival rates between women and men

#Female:
women = train_data.loc[train_data.Sex == 'female']["Survived"]
rate_women = sum(women)/len(women)

print("percentage of survived females equals", rate_women)

#Male:
men = train_data.loc[train_data.Sex == 'male']["Survived"]
rate_men = sum(men)/len(men)

print("percentage of survived males equals", rate_men)
```

Fast conclusions: Females have higher percentages of survivng

```python
#compare the survival rates between ports of embarkation

#Cherbourg:
emb_cher = train_data.loc[train_data.Embarked == 'C']["Survived"]
rate_cher = sum(emb_cher)/len(emb_cher)

print("percentage of survived people who embarked in Cherbourg", rate_cher)


#Queenstown:
emb_qu = train_data.loc[train_data.Embarked == 'Q']["Survived"]
rate_qu = sum(emb_qu)/len(emb_qu)

print("percentage of survived people who embarked in Queenstown", rate_qu)

#Southampton:
emb_so = train_data.loc[train_data.Embarked == 'S']["Survived"]
rate_so = sum(emb_so)/len(emb_so)

print("percentage of survived people who embarked in Southampton", rate_so)
```

```python
#impute Nulls for Fare by replacing them with the fare median 
train_data.Fare = train_data.Fare.fillna(train_data.Fare.median())
test_data.Fare = train_data.Fare.fillna(train_data.Fare.median())
```

```python
#impute Nulls for Pclass by replacing them with the Pclass median 
train_data.Pclass = train_data.Pclass.fillna(train_data.Pclass.median())
test_data.Pclass = train_data.Pclass.fillna(train_data.Pclass.median())
```

```python
#impute Nulls for Age by replacing them with the Age median 
train_data.Age = train_data.Age.fillna(train_data.Age.median())
test_data.Age = train_data.Age.fillna(train_data.Age.median())
```

```python
#import packages for Random Forest Classifier
import matplotlib.pyplot as plt

from collections import OrderedDict
from sklearn.datasets import make_classification

from sklearn.ensemble import RandomForestClassifier

y = train_data["Survived"]

#choosing the features for building the model
features = ["Pclass", "Sex", "Fare", "Embarked", "Age"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

#building the tree
model = RandomForestClassifier(n_estimators = 100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('my_submission.csv', index=False)
print("Your submission was successfully saved!")
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(train_data, train_data["Sex"]):
    strat_train_set = train_data.loc[train_index]
    strat_test_set = train_data.loc[test_index]
```

```python
features = ["Pclass", "Sex", "Fare", "Embarked", "Age"]  
X_train = strat_train_set[features]
y_train = strat_train_set["Survived"]
X_validation = strat_test_set[features]
y_validation = strat_test_set["Survived"]
```
